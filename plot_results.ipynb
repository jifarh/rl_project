```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_results(max_time_step):
    
    a2c = pd.read_csv(f"result/{max_time_step}/A2C.csv")
    dqn = pd.read_csv(f"result/{max_time_step}/DQN.csv")
    ppo = pd.read_csv(f"result/{max_time_step}/PPO.csv")
    
    fig, ax1 = plt.subplots(figsize=(10, 6))
    
    color = 'tab:red'
    ax1.set_xlabel('Step')
    ax1.set_ylabel('Rewards')
    ax1.plot(ppo['Step'], ppo['Value'], color=color)

    color = 'tab:blue'
    ax1.plot(dqn['Step'], dqn['Value'], color=color)
    color = 'tab:green'
    ax1.plot(a2c['Step'], a2c['Value'], color=color)
    
    plt.legend(['PPO', 'DQN', 'A2C'])
    plt.show()
```

```python
plot_results("1e6")
```

```python
plot_results("8e6")
```
